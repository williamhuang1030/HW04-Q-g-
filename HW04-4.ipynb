```python
import os
import pandas as pd
import numpy as np
import math
list = []
namelist = []
valuelist = []
positive = 0 
negative = 0
sum_pos = 0
sum_neg = 0
rankinglist = []
for file in os.listdir("C:\\Users\\User\\Desktop\\ETFF"):
    name = "C:\\Users\\User\\Desktop\\ETFF"+"\\"+file
    list.append(name)
    namelist.append(file.split(".")[0])
def omega(r):
    rf = math.pow((1+0.02822), 1/12)-1
    if r >= 0:
        ans = r - rf
    else:
        ans = r + rf
    return ans

for i in list:
    rw=[]
    df = pd.read_csv(i)
    df = df.replace(r'^\s*$', np.nan, regex=True)
    for j in df["return_month"]:
        if j != "NaN":
            rw.append(omega(j))
        else: 
            rw.append(0)
    df["return_month1"] = rw
    for i in df['return_month1']:
        if i >0:
            positive += i
            sum_pos += 1
        else:
            negative -= i
            sum_neg +=1
    valuelist.append(sum_pos/sum_neg)
c = {'ETF_symbol':namelist,'Value':valuelist}
new = pd.DataFrame(c)
result = new.sort_values(by=['Value'], ascending=False)
result1= result.iloc[1:]
result1
```
